```python
countries = pd.read_csv("countries-of-the-world.csv")
print(countries.head())
```

```python
gdp = countries["GDP ($ per capita)"]
print(gdp.head())
```

```python
phones = countries["Phones (per 1000)"]
print(phones.head())
```

```python
percent_literate = countries["Literacy (%)"]
print(percent_literate.head())
```

```python
sns.scatterplot(x= gdp, y=phones)
plt.show()
```

```python
sns.scatterplot(x=gdp, y=percent_literate)
plt.show()
```

```python
region = countries["Region"]
print(region.head())
```

```python
sns.countplot(y = region)
plt.show()
```

```python
student = pd.read_csv("student-alcohol-consumption.csv")
print(student.head())
```

```python
G3 = student["G3"]
location = student["location"]
sns.scatterplot(x="absences", y="G3", data=student, hue="location")
plt.show()
```

```python
palette_colors = {"Rural":"green", "Urban":"blue"}

sns.countplot(x="school", data= student, hue="location", palette = palette_colors)
plt.show()
```
