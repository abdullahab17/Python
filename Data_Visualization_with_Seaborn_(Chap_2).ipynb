```python
student = pd.read_csv("student-alcohol-consumption.csv")
print(student.head())
```

```python
sns.relplot(x="absences", y="G3", data=student, kind= "scatter", col="study_time")
plt.show()
```

```python
sns.relplot(x="G1", y="G3", data=student,
            kind="scatter", col="schoolsup",
            row = "famsup", col_order=["yes","no"],
            row_order=["yes","no"])
plt.show()
```

```python
mpg = pd.read_csv("mpg.csv")
print(mpg.head())
```

```python
sns.relplot(x="horsepower", y="mpg", data=mpg, kind="scatter",  size="cylinders", hue="cylinders")
plt.show()
```

```python
sns.relplot(x="acceleration", y="mpg", data=mpg, kind="scatter", hue="origin", style="origin")
plt.show()
```

```python
sns.relplot(x="model_year", y="mpg", data=mpg, kind="line")
plt.show()
```

```python
sns.relplot(x="model_year", y="mpg", data=mpg, kind="line",ci="sd")
plt.show()
```

```python
sns.relplot(x="model_year", y="mpg", data=mpg, kind="line", ci=None, hue="origin", style="origin", dashes=False, markers=True)
plt.show()
```
